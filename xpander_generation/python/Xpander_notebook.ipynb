```python
import networkx as nx
import math
import random
f = open('out.txt_mat','r')
lines = f.readlines()
G=nx.Graph()
for i, line in enumerate(lines):    
    vertices = line.split(" ")
    G.add_edge(int(vertices[0]),int(vertices[1]),capacity = 1)
```
